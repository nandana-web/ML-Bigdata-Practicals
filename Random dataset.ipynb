Consider the following 200 point X=np.dot(rng.rand(2,2),rng.randn(2,200)).T. Plot the data and using the PCA estimator reduce the 2D data into 1D. Then show the shape of the original data and the reduced data.

```python
import numpy as np
import pandas as pd

rng=np.random.RandomState(1)
X=np.dot(rng.rand(2,2),rng.randn(2,200)).T
X
df=pd.DataFrame(X,columns=["A","B"])
df
```

```python
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
#Using PCA estimator reduce the data into one dimensional.
model=PCA(n_components=1)
model.fit(df)
```

```python
df_1D=model.transform(df)
df_1D
```

```python
sns.set()
```

```python
df["PCA1"]=df_1D[:,0]
df
```

```python
#Plotting the original data
sns.lmplot(x="A",y="B",data=df,fit_reg=False)
```

```python
#Showing the shape of the original data.
df.shape
```

```python
#Showing the shape of reduced data.
df_1D.shape
```
