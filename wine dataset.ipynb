Load the dataset wine.csv and using PCA estimator reduce the data to one dimensional. Show the components and explained variance ratio. Show the shape of reduced data and original data.

```python
import numpy as np
import pandas as pd
import seaborn as sns

#Loading the wine dataset.
from sklearn.datasets import load_wine

wine=load_wine()
df=pd.DataFrame(wine["data"],columns=[wine["feature_names"]])
df.head()
```

```python
#Using PCA estimator reduce the data to one dimensional.
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=1)
model.fit(df)
```

```python
df_1D=model.transform(df)
df_1D
```

```python
df["PCA1"]=df_1D[:,0]
df
```

```python
#Explained variance(eigen values).
print(model.explained_variance_)
```

```python
#Components.
print(model.components_)
```
